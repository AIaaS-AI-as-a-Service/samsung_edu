## 문제 3. 앙상블 모델에 대한 설명으로 틀린 것을 고르시오.
1. 앙상블 모델은 크게 배깅 방법과 부스팅 방법으로 나뉘며 큰 차이점은 배깅의 경우 병렬학습을 하며 부스팅의 경우 순차학습을 한다는 점이다.
2. 배깅 방법과 부스팅 방법은 둘 다 랜덤 샘플링(복원 추출)을 통해 모델 학습을 수행한다.
3. 배깅 방법의 대표 모델인 랜덤포레스트는 out-of-bag(OOB) error를 구하여 성능 평가를 한다. 이 때 OOB는 BootStrap에 포함된 데이터를 의미한다.
4. 부스팅 방법은 잘못 분류된 데이터에 가중치를 주어 더 잘 학습하도록 한다.


## 문제 4.  아래의 그림은 Gradient Boosting Algorithm의 weak learner의 학습과정으로, Boosting 방법 중 대표적인 모델인 Gradient Boosting Algorithm에 대한 설명으로 틀린 것을 고르시오.



1. Gradient Boosting 방법은 잔차(Residual)을 weak tree가 예측하는 방식으로 학습된다.
2. 잔차(Residual)는 loss를 squared error로 설정하였을 때, negative gradient라 볼 수 있다.
3. 그림에서 tree1, tree2, tree3을 weak learner, 이를 결합한 모델을 strong learner라고 한다.
4. Weak tree를 여러 개 만들어 학습하기 때문에 과적합을 방지할 수 있다.

## 문제 5. Boosting 기법 중 대표적인 모델인 AdaBoost에 대한 설명으로 틀린 것을 고르시오.
1. AdaBoost는 여러 개의 stump(노드 하나에 두 개의 leaf를 가진 tree)로 구성되어 있다.
2. 분류한 sample에 대해서는 다른 weak learner를 만들 때, 가중치를 더 높여서 분류하게 한다.
3. Stump마다 gini index를 구하여 weight를 부여하는데 weak learner가 잘 분류할수록 gini index가 높아진다.
4. AdaBoost에서 어떤 stump의 amount of say(최종 예측에 기여하는 정도)가 크다는 것은 total error가 작다는 것이다.

## 문제 6. 앙상블 기법 중 하나인 XGBoost에 대한 설명으로 틀린 것을 고르시오.
1. GBM의 과적합 문제 및 학습속도 개선불가의 단점을 보완하고자 소개된 알고리즘이다.
2. XGBoost는 일반화 성능과 학습 효율을 향상시켜준다.
3. 데이터 저장 연산을 최소화하기 위한 block이라는 in-memory unit을 활용하여 block 단위로 재사용한다.
4. 개별 thread 마다 internal buffer를 배정해여 그리디서치의 속도 저하를 방지한다.
