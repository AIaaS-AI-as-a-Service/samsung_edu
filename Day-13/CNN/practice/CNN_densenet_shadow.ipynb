# **DenseNet 실습 (CIFAR-10 dataset)**

### **DenseNet**
- 각 레이어가 이전 모든 레이어의 출력을 입력으로 받아 특징을 공유하는 구조를 가진 신경망
- 레이어 간의 직접 연결로 매우 효율적인 파라미터 사용과 신경망이 깊어져도 학습이 잘 진행되도록 구현
- 이미지 분류, 객체 검출 등 다양한 비전 작업에 강력하며, 비교적 적은 파라미터로 높은 성능을 달성함

### **CIFAR-10 데이터 셋**
- 10개의 클래스(비행기, 자동차, 새, 고양이, 사슴, 개, 개구리, 말, 배, 트럭)로 구성된 컬러 이미지 데이터셋
- 32x32 크기의 작은 이미지 60,000장(훈련: 50,000개, 테스트: 10,000개)으로 이루어져 있으며, 딥러닝 모델 실습에 자주 사용
- 이미지 분류, CNN 학습, 전이 학습 실험 등 다양한 컴퓨터 비전 작업에 적합한 데이터셋
