#### Overfitting 발생으로 validation 곡선이 수렴성이 약하게 나타남 ####

#### Overfitting(과적합)? ###
- 모델이 학습 데이터에 과도하게 맞춰져, 새로운 데이터 혹은 테스트 데이터에 일반화되지 못하는 현상
- 원인 : 학습 데이터가 작거나 불균형하거나, 모델이 지나치게 복잡한 경우 발생
- 증상 : 학습 데이터에서 높은 정확도가 산출되지만, 테스트 데이터에서는 성능이 급격히 저하됨
- 해결책 :
    - Pooling layer 추가
    - Dropout 기법 사용
